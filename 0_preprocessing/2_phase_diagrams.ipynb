# Plots of phase diagrams

```python
import os
import numpy as np
import matplotlib.pyplot as plt
#ax.fill_between(x, y1, y2, where=y2 >= y1, facecolor='green', interpolate=True)
```

### Generating phase diagrams dir

```python
PHASE_DIAGRAMS_FIGURES_DIR = "/home/rio/ml_topological_phases_in_real_space/paper/phase_diagrams"
if not os.path.isdir(PHASE_DIAGRAMS_FIGURES_DIR):
    os.mkdir(PHASE_DIAGRAMS_FIGURES_DIR)
```

```python
x = np.linspace(-2,2,1000)
y = np.linspace(-2,2,1000)
#y = -np.linspace(-2,2,1000)
#bool1 = y1 > y2
#bool2 = -y1 < -y2
#bool1 = np.
figure1a = plt.figure(figsize = (12,12))
plt.fill_betweenx(x,y,-y,where = None, color = "red", label = 0)
plt.fill_between(x,y,-y,where = None, color = "blue", label = 1)
plt.plot(x,y,color = "white")
plt.plot(x,-y, color = "white")
plt.xlim((-2,2))
plt.ylim((-2,2))
plt.title("Phase diagram - SSH 1", fontsize=36)
plt.xlabel("$t_2$", fontsize = 48)
plt.ylabel("$t_1$", fontsize = 48)
plt.legend(loc="upper right", fontsize = 32, framealpha = 0.8)
#plt.legend(loc="upper right", framealpha = 1)
plt.yticks(fontsize = 28)
plt.xticks(fontsize= 28)
plt.tight_layout()
fname = os.path.join(PHASE_DIAGRAMS_FIGURES_DIR,"ssh1.png")
plt.savefig(fname = fname)
#plt.fill_between(x,-y,where = None, color = "red")

#plt.fill_between(x,y2,where = y2 > y1,color = "blue")
#plt.fill_between(x,y2,where = y2 > y2,color = "blue")
#plt.fill_between(x,y2,where = y2 > y2,color = "blue")
#plt.fill_between(x,y2,color = "blue")
```

```python
figure1b = plt.figure(figsize=(12,12))
### red area
x_red = [-1, 0.5, 2]
y_red = [2, 0.5, 2]
plt.fill(x_red,y_red,color = "red", label = -1)
x_red = [-2, -1, 0]
y_red = [-2, -1, -2]
plt.fill(x_red,y_red,color = "red")

### orange area
x_orange = [-1, 0, 2, 2, 0.5]
y_orange = [-1, -2, -2, -1, 0.5]
plt.fill(x_orange,y_orange,color = "orange", label = 0)

### Green area
x_green = [-2, -1, 0.5, -1, -2]
y_green = [0, -1, 0.5, 2, 2]
plt.fill(x_green,y_green,color = "green", label = 1)

### Blue area
x_blue = [-2, -1, -2]
y_blue = [-2, -1, 0]
plt.fill(x_blue,y_blue,color = "blue", label = 2)
x_blue = [0.5, 2, 2]
y_blue = [0.5, -1, 2]
plt.fill(x_blue,y_blue,color = "blue")

### plotting lines
x = np.linspace(-2,2,1000)
y1 = np.linspace(-2,2,1000)
y2 = -x + 1
y3 = -x - 2
plt.plot(x,y1,color = "white")
plt.plot(x,y2,color = "white")
plt.plot(x,y3,color = "white")
plt.xlim((-2,2))
plt.ylim((-2,2))
plt.title("Phase diagram - SSH 2", fontsize=36)
plt.xlabel("$t_2$", fontsize = 48)
plt.ylabel("$t_1$", fontsize = 48)
plt.legend(loc="upper right", fontsize = 32, framealpha = 0.8)
plt.yticks(fontsize=28)
plt.xticks(fontsize=28)
plt.tight_layout()
fname = os.path.join(PHASE_DIAGRAMS_FIGURES_DIR,"ssh2.png")
plt.savefig(fname = fname)
```
